```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

pd.options.mode.chained_assignment = None
```

```python
DATA_PATH = r'C:\Users\User\[redacted].xlsx'

df = pd.read_excel(open(DATA_PATH, 'rb'),
              sheet_name='Data')  

results = pd.DataFrame()

df.head()
```

```python
def calculate_OLS(X, y, moderators=None, model_name='model'):
    for moderator in moderators:
        X[moderator]=X[moderators[moderator]]*X['Gender']
    for column in X:
        X[column] = X[column].fillna(0)
    X1 = sm.add_constant(X)
    est = sm.OLS(y, X1)
    est2 = est.fit()
    #print(est2.summary())
    global results
    results = pd.concat((results, est2.params.rename(model_name)), axis=1)
```

```python
# X: 'Gender', 'Knowledge', 'Connections', 'Satisfaction'
# y: 'Income_change'
# moderation: 'Knowledge'

calculate_OLS(df[['Gender', 'Knowledge', 'Connections', 'Satisfaction']], 
              df['Income_change'], 
              {'moderator_kno': 'Knowledge'}, 'model_3')
```
